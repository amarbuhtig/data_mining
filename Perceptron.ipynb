```python
from __future__ import division
import random
import tensor_calculation as t
import tensor_advanced as ta
```

```python
def percep (w, x):
    if t.list_contrac(w, x+[1])>0:
        return 1
    else:
        return -1
```

```python
def train_percep(data):
    data0= [ [d[0], 1] for d in data if d[1] == [1,0]]
    data1= [ [d[0], -1] for d in data if d[1] == [0,1]]
    dim = len(data[0])
    w = ta.gene_random_tensor([dim+1], (-1,1))
    while True:
        w_old = w
        for d in data0+data1:
            w = t.list_add(t.list_nu(d[0]+[1],(d[1]- percep(w, d[0]))), w)
        if w == w_old:
            break
    return w
```

```python
data_k=ta.gene_random_tensor([1000,2], (-1, 1))
data =[]
for dk in data_k:
    if (dk[0]-0.5)**2+ dk[1]**2<0.5**2:
        data.append([dk, [1,0]])
    elif(dk[0]+0.5)**2+ dk[1]**2<0.5**2:
        data.append([dk, [0,1]])
```

```python
train_percep(data)
```

```python
def dot(v,u):
    return sum(i*j for (i,j) in zip(v,u))
```

```python
def add(v,u):
    return [i+j for (i,j) in zip(v,u)]
```

```python
def nu(a, u):
    return [a*i for i in u ]
```

```python
def vec_sum (v):
    return reduce(add, v)
```

```python
def dist(v,u):
    return dot(add(v, nu(-1,u)), add(v, nu(-1,u)))**0.5
```

```python
def mat_sum(m):
    return sum(sum(i) for i in m)
```

```python
def mat_len(c, data):
    return [[dist(i,j) for i in c] for j in data]
```

```python
def re_rang(c, data):
    l = mat_len(c, data)
    #计算中心跟每个数据的距离储存在一个矩阵里
    l_info = [i.index(min(i)) for i in l]
    #计算找出每个数据最近的一个中心，把其对应的编号储存在一个list
    data_sort =  [[] for i in range(len(c))]
    for (i,j) in zip(data, l_info):
        data_sort[j].append(i)
    #根据这个编号信息重新组织数据，把对应于一个中心的数据都放在一起
    return data_sort
    
```

```python
def km (data , k):
    c_list=[]
    #储存多组中心
    for i in range(len(data)//100+5):
    #中心的数目跟数据的长度有关
        c = random.sample(data, k)
        #随机选择几个数据作为中心
        while True:
            data_sort = re_rang(c, data)
            c_new = [nu(1/max(len(i),1), vec_sum(i))  for i in data_sort]
            #计算出新的中心
            if max([dist(i,j)/max(dot(i,i)**0.5, 0.00001)  for (i,j) in zip(c, c_new)]) < 000000.1:
                break
            #如果中心变化不大就停止循环
            c = c_new
        c_list.append(c)
        #将数据储存
    c_list_l = [ mat_sum( mat_len(i, data) ) for i in c_list]
    #判断中心的好坏
    re = c_list_l.index(min(c_list_l))
    return c_list[re], re_rang(c_list[re], data) 
```
